# Problem 3:

# The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information;


```python
x = np.random.normal(loc=5, scale = 1.5, size=1000)
x = np.sort(x)

plt.figure(figsize=(7, 5))

plt.plot(x, norm.pdf(x,loc=5,scale=1.5))
plt.xticks(range(0,10,1))
plt.show()
```

## 3.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

```python
round(norm.cdf(3.17,5,1.5),4)
```

```python
plt.plot(x, norm.pdf(x,loc=5,scale=1.5))
plt.xticks(range(0,10,1))
plt.axvline(x=3.17, c="purple")
x1 = np.linspace(x.min(), 3.17, 50)
plt.fill_between(x1, norm.pdf(x1, 5, 1.5), color="violet")
plt.show()
```

### Step 1: Defining null and alternate hypotheses. 

 Null hypothesis states that the mean hardness $ \mu $ is greater than or equal to 150

 Alternate hypothesis states that mean hardness $ \mu $ is less than 150

$ H_0 $: $\mu $ $\geq$ 150

$H_a $: $\mu $ < 150

### Step 2: Decide the significance level

Here we select $\alpha$ = 0.05.

```python
print("The sample size for this problem is",len(df))
```

### Step 3: Identify  and calculate the test statistic and p value

We do not know the population standard deviation and n = 75. So we use the t distribution and the $t_{STAT}$ test statistic.
## one tail
This is a one sample, left tailed test.

```python
from scipy.stats import ttest_1samp
```

```python
t_test, p_value= ttest_1samp(df['Unpolished '],popmean=150)
```

```python
print(f'The value of the test statistic is {round(t_test,5)}')
print(f"p-value is {round(p_value/2,10)}")
```

```python
p_value/2 # since this is a one tailed test.
#this is the value we get from t.cdf(t_test,df=74)
```

### Step 5:  Decide to reject or accept null hypothesis

```python
alpha_level = 0.05

if p_value/2 < alpha_level:
    print('We have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
```

## 5.2 Is the mean hardness of the polished and unpolished stones the same?

### Step 1: Defining null and alternate hypotheses.


Null hypothesis states that the mean hardness of polished stones is equal to the mean hardness of unpolished stones.

Alternate hypothesis states that mean hardness of polished stones is not equal to the mean hardness of unpolished stones.


$H_0$: $\mu_{polished}$ = $\mu_{unpolished}$

$H_a$: $\mu_{polished}$ $\neq$ $\mu_{unpolished}$


### Step 2: Level of significance

 $\alpha$ = 0.05

### Step 3: Identify and calculate test statistic and p- value.


This is a two tailed test.

Since the two samples are independent, we will use the independent 2 sample t-test for the hypothesis testing.

```python
from scipy.stats import ttest_ind
```

```python
t_stat, p_value = ttest_ind(df['Unpolished '],df['Treated and Polished'])
```

### Step 4: Comparing p-value and $\alpha$

```python
print(f"The value of test statistic is {round(t_stat,5)}")
print(f'The p-value is {round(p_value,8)}')
```

```python
alpha_level = 0.05

if p_value < alpha_level:
    print('We have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
```


#### Hence, we can say that the mean hardness of polished and unpolished stones are not same. 

# Problem 6:

 

# Aquarius health club, one of the largest and most popular cross-fit gyms in the country has been advertising a rigorous program for body conditioning. The program is considered successful if the candidate is able to do more than 5 push-ups, as compared to when he/she enrolled in the program. Using the sample data provided can you conclude whether the program is successful? (Consider the level of Significance as 5%)

Note that this is a problem of the paired-t-test. Since the claim is that the training will make a difference of more than 5, the null and alternative hypotheses must be formed accordingly.

```python
aq=pd.read_csv('Aquarius_gym.csv')
```

```python
aq.head()
```

```python
#since the sr. no column is of no use to use, we are dropping this column. 
aq.drop('Sr no.', axis=1, inplace=True)
```

```python
aq
```

```python
aq.isnull().sum()
```

```python
aq.describe().T
```

```python
aq.shape
```

```python
n=100
```

###  Step 1: Defining null and alternate hypotheses.
Null hypothesis states that the candidate is unable to do more than 5 push-ups, as compared to when he/she enrolled in the program.

Alternate hypothesis states that the candidate is able to do more than 5 push-ups, as compared to when he/she enrolled in the program.

Let $\mu_a $ and $\mu_b$ be the mean pushups after and before enrolling to the program respectively. 


$H_0$ : $\mu_a$ $\leq$ $\mu_b$ + 5  
$H_A$ : $\mu_a$ > $\mu_b$ + 5


OR  


$H_0$ : $\mu_a$ - $\mu_b$ $\leq$  5

$H_A$ : $\mu_a$ - $\mu_b$ >  5


### Step 2: Level of significance

Given that the level of significance, $\alpha$ = 0.05

### Step 3: Identify and calculate test statistic and p-value. 

This is a right tailed test. Since the given two samples are paired, we will use paired t-test for this testing. 

```python
from scipy.stats import ttest_rel
```

```python
t_statistic, p_value = ttest_rel(aq['Before'], aq['After'])
```

```python
print(f"The value of test statistic is {round(t_statistic,5)}")
print(f"p-value is {round(p_value/2,45)}") #since this is a one tailed test, we divide the p-value by 2. 
```

### Step 4: Decide to reject or fail to reject null hypothesis. 

```python
alpha_level = 0.05

if p_value < alpha_level:
    print('We have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternative hypothesis')
```

Hence we can conclude that a candidate is infact able to do 5 more push ups,  as compared to when he/she enrolled in the program. 

#### We can claim that the program is a success.
